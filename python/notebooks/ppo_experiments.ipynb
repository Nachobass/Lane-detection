```python
import socket
import json

# Crear conexión
#sock = socket.create_connection(("localhost", 5555))
# create socket
sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
sock.connect(("localhost", 5555))
# Mensaje como dict
msg = {"cmd": "reset"}

# Convertir a JSON string
msg_str = json.dumps(msg, separators=(",", ":"))  # {"cmd":"reset"}

# Convertir a bytes y añadir newline
msg_bytes = (msg_str + "\n").encode("utf-8")

# Enviar
sock.sendall(msg_bytes)

# Leer respuesta
resp_bytes = sock.recv(4096)  # máximo 4096 bytes
resp_str = resp_bytes.decode("utf-8")
print("Respuesta:", resp_str)

sock.close()



    # def reset(self, *, seed: Optional[int] = None, options: Optional[dict] = None) -> Tuple[np.ndarray, dict]:
    #     super().reset(seed=seed)
    #     self._connect()
    #     data = self._send({"cmd": "reset"})
    #     obs = np.array(data["obs"], dtype=np.float32)
    #     self._ensure_obs_space(obs)
    #     return obs, {}

# import os
# import numpy as np
# import matplotlib.pyplot as plt
# from stable_baselines3 import PPO
# import sys, os
# sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
# from rl_bridge.envs import make_env, UnityLaneEnv

# HOST = "127.0.0.1"
# PORT = 5555
# LOGDIR = "runs/ppo_lane"
# TIMESTEPS = 50_000

# os.makedirs(LOGDIR, exist_ok=True)
# import socket
# env = UnityLaneEnv(host=HOST, port=PORT) #make_env(HOST, PORT)
# obs, _ = env.reset()
# print("Obs shape:", obs.shape)

# model = PPO(
#     "MlpPolicy", env, verbose=1, tensorboard_log=LOGDIR,
#     n_steps=1024, batch_size=256, gae_lambda=0.95, gamma=0.995, n_epochs=10,
#     learning_rate=3e-4, clip_range=0.2,
# )

# rewards = []

# def callback(_locals, _globals):
#     # placeholder; SB3 callbacks differ here when using learn()
#     return True

# model.learn(total_timesteps=TIMESTEPS)

# env.close()
# print("Training finished.")


import os
import sys
import numpy as np
import torch
from torch.utils.tensorboard import SummaryWriter

# Add parent directory to path to import PPO module
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from rl_bridge.envs import UnityLaneEnv
from PPO.ppo_clip import PPOClip, PPOConfig

HOST = "127.0.0.1"
PORT = 5555
LOGDIR = "runs/ppo_lane"
TIMESTEPS = 50_000
# TIMESTEPS = 100

os.makedirs(LOGDIR, exist_ok=True)

# Crear entorno y conectarlo a Unity
env = UnityLaneEnv(host=HOST, port=PORT)
obs, _ = env.reset()
print("Obs shape:", obs.shape)

# Give Unity time to initialize after reset
import time
time.sleep(0.1)  # Small delay to ensure Unity is ready
print("Unity environment ready")

# Setup device and tensorboard
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using {device} device")
writer = SummaryWriter(LOGDIR)

# Get observation and action dimensions
obs_dim = int(obs.shape[0])
act_dim = 2  # turn, throttle in [-1, 1]

# Create PPO config with same hyperparameters as before
cfg = PPOConfig(
    n_steps=1024,
    batch_size=256,
    gae_lambda=0.95,
    gamma=0.995,
    n_epochs=10,
    learning_rate=3e-4,
    clip_range=0.2,
)

# Crear modelo PPO (custom implementation)
agent = PPOClip(obs_dim=obs_dim, act_dim=act_dim, config=cfg, device=device)

# Test Unity step response before training (only once)
print("Testing Unity step response before training...")
try:
    test_obs, test_reward, test_done, _, _ = env.step(np.array([0.0, 0.0], dtype=np.float32))
    print(f"Test step successful! Obs: {test_obs.shape}, Reward: {test_reward}")
    obs = test_obs  # Use test observation
except RuntimeError as e:
    print(f"ERROR: Unity is not responding to step commands: {e}")
    print("Please check:")
    print("1. Unity simulator is running")
    print("2. Unity is listening on port 5555")
    print("3. CarController and Sensors are properly set up in Unity")
    raise

# Training loop
episode_return = 0.0
episode_len = 0
total_steps = 0

while total_steps < TIMESTEPS:
    # Collect rollout of length cfg.n_steps
    for _ in range(cfg.n_steps):
        obs_tensor = torch.tensor(obs, dtype=torch.float32, device=device)
        with torch.no_grad():
            action, logp, value = agent.select_action(obs_tensor)
        # Ensure action is a 1D numpy array with shape (2,)
        action_np = action.numpy().flatten()
        if action_np.shape != (2,):
            action_np = action_np.reshape(2)
        # Ensure it's float32 and clipped to valid range
        action_np = np.clip(action_np, -1.0, 1.0).astype(np.float32)
        
        next_obs, reward, terminated, truncated, _ = env.step(action_np)
        
        agent.buffer.add(obs, action_np, logp, float(reward), bool(terminated), float(value))
        
        episode_return += float(reward)
        episode_len += 1
        total_steps += 1
        obs = next_obs
        
        if terminated or truncated:
            writer.add_scalar("rollout/episode_return", episode_return, total_steps)
            writer.add_scalar("rollout/episode_len", episode_len, total_steps)
            obs, _ = env.reset()
            time.sleep(0.1)  # Give Unity time after reset
            episode_return = 0.0
            episode_len = 0
    
    # Bootstrap value for GAE
    with torch.no_grad():
        last_value = agent.value(torch.tensor(obs, dtype=torch.float32, device=device).unsqueeze(0)).item()
    batch = agent.buffer.get(last_value=last_value, gamma=cfg.gamma, lam=cfg.gae_lambda)
    
    # Gradient updates
    metrics = agent.update(batch)
    for k, v in metrics.items():
        writer.add_scalar(f"train/{k}", v, total_steps)

env.close()
writer.close()
print("Training finished.")
```
